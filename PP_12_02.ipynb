## Show the male female survive

* Task 1: Use the scattered plot with (X: Age), (Y: Fare) and provide different colors for male survived and female survived.

* Task 2: Create 2 subplots for male and female passengers who survived. Visualize the age histogram for each plot.

```python
#show the male and female survived graph
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = sns.load_dataset("titanic")

# Filter only survivors
df_survived = df[df['survived'] == 1]

# Define colors for male and female survivors
colors = {'male': 'blue', 'female': 'red'}

# Create scatter plot
plt.figure(figsize=(10, 6))
for gender, color in colors.items():
    subset = df_survived[df_survived['sex'] == gender]
    plt.scatter(subset['age'], subset['fare'], c=color, label=gender, alpha=0.6, edgecolors='k')

# Labels and title
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Age vs Fare for Survivors")
plt.legend(title="Sex")
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = sns.load_dataset("titanic")

# Filter only survivors
df_survived = df[df['survived'] == 1]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)

# Plot histograms for male and female survivors
for i, gender in enumerate(['male', 'female']):
    df_survived[df_survived['sex'] == gender]['age'].plot(kind='hist', bins=15, ax=axes[i], color='blue' if gender == 'male' else 'red', edgecolor='black', alpha=0.7)
    axes[i].set_title(f'{gender.capitalize()} Survivors')
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Count')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
